```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('../input/nba-regular-season-stats-20182019/nbastats2018-2019.csv')
data.head()
```

```python
data = data[(data['G'] >= 30) & (data['MP'] >= 10)]
data = data.drop(['Salary', 'Team', 'Age', 'G', 'MP', 'VORP', 'BPM'], 1)
data['FG3%'] = data['FG3%'].fillna(0)
```

```python
plt.figure(figsize=(20, 15))
sns.heatmap(data=data.corr(), annot=True)
```

```python
model_data = data.drop('Name', 1)
num_c = 12
model = KMeans(n_clusters = num_c, random_state = 0).fit(model_data)
labels = model.labels_
```

```python
player_groups = pd.DataFrame(data = {'Name': data['Name'], 'Group': labels})
groups = {}

for group in range(num_c):
    groups[f"Group {group+1}"] = list(player_groups.query(f"Group == {group}")['Name'])
    
for group in groups:
    print(f"{group}:\n")
    print(groups[group], "\n\n\n")
```
